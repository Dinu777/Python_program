```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import warnings

warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv('kr-vs-kp.csv', header=0, sep=',')
```

```python
dataset.head()
```

```python
dataset = dataset.dropna()
print (dataset.shape)
```

```python
print(list(dataset.columns))
```

```python
data = pd.get_dummies(dataset, columns = ['f', 'f.1', 'f.2', 'f.3', 'f.4', 'f.5', 'f.6',
                                          'f.7', 'f.8', 'f.9', 'f.10', 'f.11', 'l', 'f.12', 
                                          'n', 'f.13', 'f.14', 't', 'f.15', 'f.16', 'f.17', 
                                          'f.18', 'f.19', 'f.20', 'f.21', 't.1', 'f.22', 'f.23', 
                                          'f.24', 'f.25', 'f.26', 'f.27', 'f.28', 't.2', 't.3', 'n.1'])
```

```python
data.head()
```

```python
data.columns
```

```python
X = data.iloc[:,1:]
```

```python
X.head()
```

```python
Y = data.iloc[:,0]
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
boost = AdaBoostClassifier(n_estimators=50, learning_rate=1)
model = boost.fit(X_train, Y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```

```python
cnf_matrix = metrics.confusion_matrix(Y_test, y_pred)
print(cnf_matrix)
```

```python
print(metrics.classification_report(Y_test, y_pred))
```
